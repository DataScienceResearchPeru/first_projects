```python
movie_df=pd.read_csv('2013_movies.csv')
n=len(movie_df)
movie_df.columns
```

**Challenge 1**

- Plot domestic total gross over time.

```python
#Create new column with just the release date (Is currently DateTime)
movie_df['Date_dt']=[pd.to_datetime(x).date() for x in movie_df['ReleaseDate']]
total_gross = movie_df.groupby('Date_dt')['DomesticTotalGross'].sum().reset_index()
total_gross.head()
```

```python
%pylab inline

plt.figure(figsize=(10,5))
plt.plot(total_gross['Date_dt'],total_gross['DomesticTotalGross'])

plt.title('Daily Total Gross of Top 100 movies (2013)', fontsize=15)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Billion $', fontsize=12)
```

Bin your dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean.

Title of graph should include: Mean Domestic Total Gross by Month in 2013
