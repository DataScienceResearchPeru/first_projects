### Correlation

**Correlation Heatmap**

```python
f, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(coffee_df.corr(), annot=False, linewidths=.5, cmap='YlOrRd')
```

Use only between first 12 variables (Do not include, monthly data or intercept)

```python
f, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(coffee_df.iloc[:,:13].corr(), annot=False, linewidths=.5, cmap='YlOrRd')

plt.title('Correlation HeatMap')
```

```python
sns.pairplot(coffee_df.iloc[:,:13])
```

**First MVP**

```python
coffee_df.columns[3:13]
```

```python
# Lineal Regression for EACH dependent variable against Production Values
plt.figure(figsize=(20, 30))
pos=1
for i in coffee_df.columns[3:13] :
    plt.subplot(5, 2, pos)
    sns.regplot(x=i, y='Produccion', data=coffee_df, color='purple', marker=(5, 2))
    plt.ylabel('Production')
    plt.xlabel(i)
    pos+=1

# First MVP!
plt.savefig('../graphs/initial_regplot.png')
```
